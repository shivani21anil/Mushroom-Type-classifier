So there are no missing vlaues in our data. But when you go through the data, you will find that missing values in one column is replaced with '?'.
We need to handle it.

```python
for column in data.columns:
    count = data[column][data[column]=='?'].count()
    if count!=0:
        print(column, data[column][data[column]=='?'].count())    
```

There are 2480 missing values in "stalk-root" column. Let's replace such values with numpy "nan" so that we can handle the missing values.

```python
data["stalk-root"] =data["stalk-root"].replace('?',np.nan)  
```

```python
data.isna().sum()
```

Great! now we can see the missing values in the stalk-root column. we will impute these missing values.





```python
for column in data.columns:
    print(column,data[column].nunique())    
```

Notice that "veil-type" column has only one unique category. This column won't give much information ,so let's drop it.

```python
data = data.drop(['veil-type'],axis=1)
```

Great! Let's start with imputing the missing values. We will use "autoImputer" package. It provides various techniques to impute missing values. We will use Single Imputer here.

you can read more about "autoImputer" here :

https://pypi.org/project/autoimpute/

https://readthedocs.org/projects/autoimpute/downloads/pdf/latest/

```python
impute = SingleImputer(strategy='categorical')
```

```python
impute.fit(data)
```

```python
new_data =impute.transform(data)
```

```python
new_data
```

```python
sns.countplot(new_data["stalk-root"])
```

```python
sns.countplot(data["stalk-root"])
```

```python
new_data.isna().sum()
```

Awesome! the missing values have been imputed and there are no missing values now.

Also, we did not have to encode the categorical values first before imputing as we did with KNN imputer.

Let's move ahead and encode our categorical values.

Before we impute the missing values, let's first encode the categoical values.

let's check the number of unique caegories each column has.

```python
for column in new_data.columns:
    print(column,new_data[column].nunique())
```

We will map columns "class". we are mapping class column, so we can easily return the actual value during prediction.
We will use get dumies for rest of the columns.

```python
new_data["class"] = new_data["class"].map({"'p'" : 0, "'e'" : 1})
```

```python
#col = new_data.drop(['class'],axis=1).columns
```

```python
for column in new_data.drop(['class'],axis=1).columns:
    new_data = pd.get_dummies(new_data, columns=[column])
```
